```python
import pandas as pd
#Veri setini oluşturma
data={
    'Year':[2020,2021,2022,2023,2024],
    'Sales (USD)': [50000,75000,100000,120000,15000],
    'Customers': [120,180,250,300,400]
}

df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,6))
```

```python
#satışların çizgi grafiği
plt.subplot(1,2,1)
plt.plot(df['Year'],df['Sales (USD)'],marker='o',color='Black')
plt.title('Yıllık Satışlar')
plt.xlabel('Yıl')
plt.ylabel('Satış (USD)')
plt.grid()
# Müşteri Sayısının Çubuk Grafiği
plt.subplot(1,2,2)
plt.bar(df['Year'],df['Customers'],color='orange')
plt.title('Yllık Müşteri Sayısı')
plt.xlabel('Yıl')
plt.ylabel('Müşteri Sayısı')
```
